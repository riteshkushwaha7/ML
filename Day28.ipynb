```python
x_train[['Class']] = oe.fit_transform(x_train[['Class']])
x_test[['Class']] = oe.transform(x_test[['Class']])
```

# nominal encoding

```python
# using pandas K-1 OneHotEncoding
pd.get_dummies(df,columns=['Gender','Customer Type','Type of Travel'],drop_first=True)
```

```python
# OneHotEncoding using Sklearn
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first',sparse_output=True,dtype=np.int32)
X_train_new = ohe.fit_transform(x_train[['Gender','Customer Type','Type of Travel']])
X_test_new = ohe.transform(x_test[['Gender','Customer Type','Type of Travel']])
```

```python
df.columns
```

```python
numeric_cols = ['Unnamed: 0', 'id', 'Age', 'Class', 'Flight Distance',
                'Inflight wifi service','Departure/Arrival time convenient',
                'Ease of Online booking','Gate location','Food and drink',
                'Online boarding','Seat comfort','Inflight entertainment',
                'On-board service','Leg room service','Baggage handling',
                'Checkin service','Inflight service','Cleanliness',
                'Departure Delay in Minutes','Arrival Delay in Minutes']

# One-hot feature names
ohe_feature_names = ohe.get_feature_names_out(['Gender','Customer Type','Type of Travel'])
X_train_final = pd.DataFrame(np.hstack((x_train[numeric_cols].values, X_train_new.toarray())),
                             columns=numeric_cols + list(ohe_feature_names),
                             index=x_train.index)

X_test_final = pd.DataFrame(np.hstack((x_test[numeric_cols].values, X_test_new.toarray())),
                            columns=numeric_cols + list(ohe_feature_names),
                            index=x_test.index)
```

```python
X_train_final = X_train_final.fillna(0).astype('int32')
X_test_final  = X_test_final.fillna(0).astype('int32')
```
